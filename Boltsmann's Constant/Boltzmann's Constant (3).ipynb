```python
e = 1.602176634e-19  # C

Resistors = [R1M,   R100k,   R10k,   R1k]
Temps     = [T_1M,  T_100k,  T_10k,  T_1k]   # °C
R_values  = [1e6,   1e5,     1e4,    1e3]    # Ω
V_true_ls = [V_true_1M, V_true_100k, V_true_10k, V_true_1k]
dV_true_ls= [dV_true_1M,dV_true_100k,dV_true_10k,dV_true_1k]
V_E_ls    = [V_E_1M,   V_E_100k,   V_E_10k,   V_E_1k]

kbs = []
dkbs =[]

for R, T_C, V_true, dV_true, V_E in zip(R_values, Temps, V_true_ls, dV_true_ls, V_E_ls):
    # current and ln I (no abs; sign handled by the minus)
    I   = np.abs(V_true) / R
    lnI = np.log(I)

    # error in ln I (propagate only voltage error)
    dI = np.sqrt(((1/R)*dV_true)**2 + ((V_true/R**2)*(R*0.05))**2)
    dlnI = np.sqrt(((1/I)*dI)**2)

    # weighted linear fit lnI vs V_E
    m, b, dm, db, fig = PHYS231.LinearFit(
        xData   = np.abs(V_E),
        yData   = lnI,
        yErrors = dlnI,
        xlabel  = 'V_E',
        ylabel  = 'ln I',
        xUnits  = '',
        yUnits  = ''
    )

    # temperature in K and k_B from slope
    T_K = T_C + 273.15
    kB  = e / (m * T_K)
    dkB = np.sqrt(((-e/T_K)*dm)**2)
    kbs.append(kB)
    dkbs.append(dkB)
    print(f"R = {R:.0e} Ω, T = {T_K:.2f} K, m = {m:.3g}, "
    f"k_B = ({kB:.3e} ± {dkB:.3e}) J/K"
    )
    


# kbs now holds the four k_B estimates


```
