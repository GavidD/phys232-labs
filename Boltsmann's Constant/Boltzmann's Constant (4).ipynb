```python
import numpy as np
import matplotlib.pyplot as plt

# Your dataset definition here...
datasets = [
    {"label": "1M",   "V_t": V_true_1M,   "dV_t": dV_true_1M,   "V_E": V_E_1M,   "dV_E": dV_E_1M,   "R": 1e6},
    {"label": "100k", "V_t": V_true_100k, "dV_t": dV_true_100k, "V_E": V_E_100k, "dV_E": dV_E_100k, "R": 100e3},
    {"label": "10k",  "V_t": V_true_10k,  "dV_t": dV_true_10k,  "V_E": V_E_10k,  "dV_E": dV_E_10k,  "R": 10e3},
    {"label": "1k",   "V_t": V_true_1k,   "dV_t": dV_true_1k,   "V_E": V_E_1k,   "dV_E": dV_E_1k,   "R": 1e3}
]

plt.figure(figsize=(10, 7))

# Initializing lists to collect all data points for the fit
all_ve = []
all_ln_i = []
all_dve = []
all_dln_i = []
all_labels = []

for data in datasets:
    # Cleanup and conversion
    ve  = np.array(data["V_E"],  dtype=float)
    dve = np.array(data["dV_E"], dtype=float)
    vt  = np.array(data["V_t"],  dtype=float)
    dvt = np.array(data["dV_t"], dtype=float)
    R   = data["R"]
    dR  = 0.05 * R  # 5% resistor uncertainty
    
    # Calculate I and ln(I)
    I   = vt / R
    dI  = np.abs(I) * np.sqrt((dvt / vt)**2 + (dR / R)**2)
    ln_I   = np.log(I)
    dln_I  = dI / I
    
    # Append to collective lists for fitting
    all_ve.extend(ve)
    all_ln_i.extend(ln_I)
    all_dve.extend(dve)
    all_dln_i.extend(dln_I)
    all_labels.extend([data["label"]] * len(ve))
    
    # (Optional) you could plot here, but we'll replot after outlier mask

all_ve    = np.array(all_ve)
all_ln_i  = np.array(all_ln_i)
all_dve   = np.array(all_dve)
all_dln_i = np.array(all_dln_i)
all_labels = np.array(all_labels)

# --- First fit for outlier detection ---
m0, b0 = np.polyfit(all_ve, all_ln_i, 1)
residuals0 = all_ln_i - (m0 * all_ve + b0)
sr0 = np.std(residuals0, ddof=1)

# --- 2σ outlier mask ---
mask = np.abs(residuals0) < 2 * sr0

print("Removed outliers (ROOM TEMP):")
for i in np.where(~mask)[0]:
    print(f" index {i}, label = {all_labels[i]}, V_E = {all_ve[i]:.4f}, ln(I) = {all_ln_i[i]:.4f}")

# --- Plot inliers vs outliers ---
colors = {"1M": "C0", "100k": "C1", "10k": "C2", "1k": "C3"}

for label in ["1M", "100k", "10k", "1k"]:
    keep = (all_labels == label) & mask
    out  = (all_labels == label) & (~mask)
    
    plt.errorbar(all_ve[keep], all_ln_i[keep],
                 xerr=all_dve[keep], yerr=all_dln_i[keep],
                 label=f"Resistor: {label}",
                 fmt='o', markersize=4, capsize=3, elinewidth=1,
                 alpha=0.8, color=colors[label])
    
    if np.any(out):
        plt.errorbar(all_ve[out], all_ln_i[out],
                     xerr=all_dve[out], yerr=all_dln_i[out],
                     fmt='x', markersize=8, capsize=3, elinewidth=1,
                     alpha=0.5, color=colors[label], label=f"Outlier: {label}")

# --- Final fit after outlier removal ---
x = all_ve[mask]
y = all_ln_i[mask]

m, b = np.polyfit(x, y, 1)

# --- Slope/intercept uncertainties ---
n = len(x)
y_fit_points = m * x + b
residuals = y - y_fit_points
s_yx = np.sqrt(np.sum(residuals**2) / (n - 2))
Sxx = np.sum((x - np.mean(x))**2)

dm = s_yx / np.sqrt(Sxx)
db = s_yx * np.sqrt(np.sum(x**2) / (n * Sxx))

# --- Fit line for plot ---
x_fit = np.linspace(np.min(x), np.max(x), 100)
y_fit = m * x_fit + b
plt.plot(x_fit, y_fit, color='black', linestyle='--',
         label=f"Fit Line: m = {m:.4f} ± {dm:.4f}")

# Final Formatting
plt.xlabel("$V_E$ (Volts)")
plt.ylabel("$\\ln(I)$")
plt.title("$\\ln(I)$ vs $V_E$ with Propagated Uncertainties and Linear Fit ROOM TEMP")
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)

print(f"ROOM TEMP slope m = {m:.4f} ± {dm:.4f}")
print(f"ROOM TEMP intercept b = {b:.4f} ± {db:.4f}")

# --- Temperature average + uncertainty (room temp) ---
T_vals_C_room = np.array([T_1M, T_100k, T_10k, T_1k], dtype=float)  # or room-temp-specific Ts
T_vals_K_room = T_vals_C_room + 273.15
T_avg_room = np.mean(T_vals_K_room)

# Replace 0.5 with your actual thermometer sigma if different
dT_vals_room = np.array([0.5, 0.5, 0.5, 0.5], dtype=float)
dT_avg_room = np.sqrt(np.sum(dT_vals_room**2)) / len(dT_vals_room)

# --- Boltzmann constant from room-temp data ---
kb_room = -e / (m * T_avg_room)
dkb_room = abs(kb_room) * np.sqrt((dm / m)**2 + (dT_avg_room / T_avg_room)**2)

print(f"T_avg (ROOM) = {T_avg_room:.2f} ± {dT_avg_room:.2f} K")
print(f"kb (ROOM) = ({kb_room:.4e} ± {dkb_room:.4e}) J/K")

plt.show()
```

```python
datasets = [
    {"label": "1M",   "V_t": CV_true_1M,   "dV_t": CdV_true_1M,   "V_E": CV_E_1M,   "dV_E": CdV_E_1M,   "R": 1e6},
    {"label": "100k", "V_t": CV_true_100k, "dV_t": CdV_true_100k, "V_E": CV_E_100k, "dV_E": CdV_E_100k, "R": 100e3},
    {"label": "10k",  "V_t": CV_true_10k,  "dV_t": CdV_true_10k,  "V_E": CV_E_10k,  "dV_E": CdV_E_10k,  "R": 10e3},
    {"label": "1k",   "V_t": CV_true_1k,   "dV_t": CdV_true_1k,   "V_E": CV_E_1k,   "dV_E": CdV_E_1k,   "R": 1e3}
]

plt.figure(figsize=(10, 7))

all_ve = []
all_ln_i = []
all_dve = []
all_dln_i = []
all_labels = []

for data in datasets:
    ve  = np.array(data["V_E"], dtype=float)
    dve = np.array(data["dV_E"], dtype=float)
    vt  = np.array(data["V_t"], dtype=float)
    dvt = np.array(data["dV_t"], dtype=float)
    R   = data["R"]
    dR  = 0.05 * R   # 5% resistor uncertainty

    # Current and uncertainty
    I = vt / R
    dI = I * np.sqrt((dvt / vt)**2 + (dR / R)**2)

    # ln(I) and uncertainty
    ln_I = np.log(I)
    dln_I = dI / I

    all_ve.extend(ve)
    all_ln_i.extend(ln_I)
    all_dve.extend(dve)
    all_dln_i.extend(dln_I)
    all_labels.extend([data["label"]] * len(ve))

all_ve = np.array(all_ve)
all_ln_i = np.array(all_ln_i)
all_dve = np.array(all_dve)
all_dln_i = np.array(all_dln_i)
all_labels = np.array(all_labels)

# Initial fit
m0, b0 = np.polyfit(all_ve, all_ln_i, 1)
residuals0 = all_ln_i - (m0 * all_ve + b0)
sr0 = np.std(residuals0, ddof=1)

# 2-sigma outlier removal
mask = np.abs(residuals0) < 2 * sr0

print("Removed outliers:")
for i in np.where(~mask)[0]:
    print(f" index {i}, label = {all_labels[i]}, V_E = {all_ve[i]:.4f}, ln(I) = {all_ln_i[i]:.4f}")

# Refit after removing outliers
x = all_ve[mask]
y = all_ln_i[mask]

m, b = np.polyfit(x, y, 1)

# Slope/intercept uncertainty calculation
n = len(x)
y_fit = m * x + b
residuals = y - y_fit
s_yx = np.sqrt(np.sum(residuals**2) / (n - 2))
Sxx = np.sum((x - np.mean(x))**2)

dm = s_yx / np.sqrt(Sxx)
db = s_yx * np.sqrt(np.sum(x**2) / (n * Sxx))

# Plot kept points and outliers
colors = {"1M": "C0", "100k": "C1", "10k": "C2", "1k": "C3"}

for label in ["1M", "100k", "10k", "1k"]:
    idx_keep = (all_labels == label) & mask
    idx_out  = (all_labels == label) & (~mask)

    plt.errorbar(all_ve[idx_keep], all_ln_i[idx_keep],
                 xerr=all_dve[idx_keep], yerr=all_dln_i[idx_keep],
                 label=f"Resistor: {label}",
                 fmt='o', markersize=4, capsize=3, elinewidth=1,
                 alpha=0.8, color=colors[label])

    if np.any(idx_out):
        plt.errorbar(all_ve[idx_out], all_ln_i[idx_out],
                     xerr=all_dve[idx_out], yerr=all_dln_i[idx_out],
                     fmt='x', markersize=8, capsize=3, elinewidth=1,
                     alpha=0.5, color=colors[label], label=f"Outlier: {label}")

# Fit line
x_fit = np.linspace(np.min(x), np.max(x), 100)
y_fit_line = m * x_fit + b
plt.plot(x_fit, y_fit_line, color='black', linestyle='--',
         label=f"Fit Line: m = {m:.4f} ± {dm:.4f}")

plt.xlabel("$V_E$ (Volts)")
plt.ylabel("$\\ln(I)$")
plt.title("$\\ln(I)$ vs $V_E$ with Propagated Uncertainties and Linear Fit COLD DATA")
plt.legend()
plt.grid(True, which='both', linestyle='--', alpha=0.5)

print(f"Slope m = {m:.4f} ± {dm:.4f}")
print(f"Intercept b = {b:.4f} ± {db:.4f}")

# --- Temperature handling ---
# If your T values are in Celsius:
T_vals_C = np.array([T_1M, T_100k, T_10k, T_1k], dtype=float)
T_vals_K = T_vals_C + 273.15

T_avg = np.mean(T_vals_K)

# If you know thermometer uncertainties, put them here.
# Example: 0.5 C each
dT_vals = np.array([0.5, 0.5, 0.5, 0.5], dtype=float)
dT_avg = np.sqrt(np.sum(dT_vals**2)) / len(dT_vals)

# Boltzmann constant
kb = -e / (m * T_avg)

# Propagate kb uncertainty from m and T
dkb = abs(kb) * np.sqrt((dm / m)**2 + (dT_avg / T_avg)**2)

print(f"T_avg = {T_avg:.2f} ± {dT_avg:.2f} K")
print(f"kb = ({kb:.4e} ± {dkb:.4e}) J/K")

plt.show()
```
